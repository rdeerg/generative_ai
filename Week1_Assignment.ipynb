```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#參數定義
a = 0 #起始半徑
b = 1 #螺線間距
theta = np.linspace(0, 6*np.pi, 500) #角度範圍(從0-6π，共繪製500個點)

"""一般畫螺線用極座標較為直觀，但常用座標系統為直角座標"""
#轉換為直角座標
r = a + b * theta #計算極座標半徑
x = r * np.cos(theta) #轉換為 x 座標
y = r * np.sin(theta) #轉換為 y 座標
```

```python
#繪製阿基米德螺線
plt.figure(figsize=(6,6)) #設定畫布大小
plt.plot(x, y, color='r', linewidth=2) #畫出螺線，顏色為紅色，寬度為2
plt.axhline(0, color='black', linewidth=0.5, linestyle="--") #加上x軸0.5寬虛線
plt.axvline(0, color='black', linewidth=0.5, linestyle="--") #加上y軸0.5寬虛線
plt.xlabel("x") #x軸座標
plt.ylabel("y") #y軸座標
plt.title("Archimedean Spiral") #標題
plt.grid() #輔助線
```
